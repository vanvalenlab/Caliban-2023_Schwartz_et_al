```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../benchmarking/DeepCell/speed.csv')
```

```python
df['tracking'] = df['ne'] + df['inf'] + df['lap']
```

```python
avg = df.groupby('batch').mean().reset_index()
```

```python
fig, ax = plt.subplots(2,2, sharey='row', sharex='col', figsize=(6,6))
ax[0,0].scatter(avg['seg'], avg['nobjects']/1000, c=avg['batch'], cmap='tab20')
ax[1,0].scatter(avg['seg'], avg['nframes'], c=avg['batch'], cmap='tab20')
ax[0,1].scatter(avg['tracking'], avg['nobjects']/1000, c=avg['batch'], cmap='tab20')
ax[1,1].scatter(avg['tracking'], avg['nframes'], c=avg['batch'], cmap='tab20')

ax[0,0].set_title('Segmentation')
ax[0,1].set_title('Tracking')

ax[0,0].set_ylabel('Number of Objects (k)')
ax[1,0].set_ylabel('Number of Frames')

ax[1,0].set_xlabel('Time (s)')
ax[1,1].set_xlabel('Time (s)')

plt.tight_layout()
```
