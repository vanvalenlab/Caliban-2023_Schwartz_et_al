```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/notebooks/benchmarking/DeepCell/speed.csv')
```

```python
df['tracking'] = df['ne'] + df['inf'] + df['lap']
```

```python
avg = df.groupby('batch').mean().reset_index()
```

```python
fig, ax = plt.subplots(2,2, sharey='row', sharex='col', figsize=(6,6))
ax[0,0].scatter(avg['nobjects']/1000, avg['seg'], c=avg['batch'], cmap='tab20')
ax[0,1].scatter(avg['nframes'], avg['seg'], c=avg['batch'], cmap='tab20')
ax[1,0].scatter(avg['nobjects']/1000, avg['tracking'], c=avg['batch'], cmap='tab20')
ax[1,1].scatter(avg['nframes'], avg['tracking'], c=avg['batch'], cmap='tab20')

# ax[0,0].set_title('Segmentation')
# ax[0,1].set_title('Tracking')

ax[1,0].set_xlabel('Number of Objects (k)')
ax[1,1].set_xlabel('Number of Frames')

ax[0,0].set_ylabel('Time (s)')
ax[1,0].set_ylabel('Time (s)')

plt.tight_layout()
plt.savefig('speed.pdf')
```
